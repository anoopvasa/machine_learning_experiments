# Flights Data Analysis

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib as mlp
%matplotlib inline
import seaborn as sns

flights_dframe = sns.load_dataset('flights')
flights_dframe.head()
```

```python
flights_df = flights_dframe.pivot('month','year','passengers')
flights_df.head()
```

```python
sns.heatmap(flights_df,annot=True,fmt='d',center=flights_df.loc['January',1955])
```

```python
yearly_flights = flights_df.sum()
yearly_flights.head()
```

```python
fig, (axis1, axis2) =plt.subplots(2,1)
years = Series(yearly_flights.index.values)
years = DataFrame(years)
flights_total = Series(yearly_flights.values)
flights = DataFrame(flights_total)
yearly_flights = pd.concat((years, flights), axis= 1)
yearly_flights.columns = ['Year','Flights']
sns.barplot('Year',y='Flights',data=yearly_flights, ax=axis1)
sns.heatmap(flights_df,cmap='Blues',ax=axis2,cbar_kws={'orientation':'horizontal'})
```
