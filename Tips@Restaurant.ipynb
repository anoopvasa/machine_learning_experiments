# Linear Regression Experiment with Seaborn

```python
#We are going to analyse an inbuilt dataset called tips that comes along with seaborn
```

```python
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
from numpy.random import randn

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tips_pect'] = (tips['tip']/tips['total_bill'])*100
sns.lmplot('total_bill','tip',tips, order=4,fit_reg = True,
           scatter_kws={'marker':'o', 'color':'indianred'}, 
           line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tips_pect',tips,hue='sex',markers=['o','+'])
```

```python
sns.lmplot('total_bill','tips_pect',tips,hue='day')
```
