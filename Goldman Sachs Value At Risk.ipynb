# Value at Risk / Monte Carlo Method

```python
days = 365
dt = 1/365
mu = rets.mean()['GS']

sigma = rets.std()['GS']

def monte_carlo(start_price, days, mu, sigma ):
    price = np.zeros(days)
    price[0] = start_price
    
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    for x in range(1,days):
        shock[x] = np.random.normal(loc=mu*dt, scale=sigma*np.sqrt(dt))
        drift[x] = mu * dt
        
        price[x] = price[x-1] + ( price[x-1] * (drift[x] + shock[x]))
        
    return price
```

```python
start_price = 152

for run in range(100) :
    plt.plot(monte_carlo(start_price, days, mu, sigma ))
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.title('Monte Carlo Analysis for Goldman Sachs')
plt.show()
```

```python
runs = 10000
sims = np.zeros(runs)
for run in range(runs):
    sims[run] = monte_carlo(start_price, days, mu, sigma)[days - 1]
    
```

```python
q = np.percentile(sims,1)
plt.hist(sims, bins = 200)
#Some info
plt.figtext(0.6,0.8, s= 'Start Price: $%.2f' %start_price)
plt.figtext(0.6,0.7, s= 'Mean Final Price: $%.2f' %sims.mean())
plt.figtext(0.6,0.6, s= 'VaR(0.99): $%.2f' %(start_price -q))
plt.figtext(0.6,0.5, s= 'Quantile(0.99): $%.2f' %float(q))
            
plt.axvline(x=q, linewidth=4, color='r')
plt.title(u"Final Price Ditribution of Goldman Sachs International after %s days" %(days), weight='bold')
```
