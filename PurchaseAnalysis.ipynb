# Analysis to interpret relationship between Purchase Amt and Page Speed of an online shopping website

```python
#Method: We would want to check the correlation and covariance of the the features above mentioned
#A strong inverse relationship would signify that faster the page loads, more is purchased
#We create dummy data to case-study this example
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.xkcd()

#Function that computes devation of vector dimensions from mean
def de_mean(x):
    mu = np.mean(x)
    return x - mu

#Function that calculates the covariance
def covariance(x1,x2):
    n = len(x1)
    return np.dot(de_mean(x1), de_mean(x2))/ (n)

#Setup random data
pageSpeeds = np.random.normal(3,1,1000)
purchaseAmounts = np.random.normal(50,10,1000)
print 'Covariance ',covariance(pageSpeeds, purchaseAmounts)

plt.xlabel('Page Speeds of Website')
plt.ylabel('Purchase Amount on Website')
plt.scatter(pageSpeeds,purchaseAmounts, color = 'green')
plt.show()
```

```python
#Clearly we can see in the above graph that both these features are not well correlated
#Lets now define an explicit relationship betweeen pageSpeeds and purchaseAmts
purchaseAmounts = np.random.normal(2,2,1000)/pageSpeeds + 40
print 'Covariance ',covariance(pageSpeeds, purchaseAmounts)
plt.xlabel('Page Speeds of Website')
plt.ylabel('Purchase Amount on Website')
plt.scatter(pageSpeeds,purchaseAmounts, color = 'red')
plt.show()
```

```python
#The above graph is a good example to express that with increasing pageSpeeds, less amount of goods is purchased
#The covariance as such is not very helpful, hence we need to calculate correlation

def correlation(x,y):
    return covariance(x,y)/x.std()/y.std()

#Get correlation of the last graph
print 'Correlation ', correlation(pageSpeeds, purchaseAmounts)
```

```python
#Alternative way to get correlation coefs
#This clearly indicated that purchaseAmounts is negetively correlated to pageSpeeds, but does not imply causation
print np.corrcoef(pageSpeeds, purchaseAmounts)
```
